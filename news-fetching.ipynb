Fetching News Sentiment

```python
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
url = 'https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers=FOREX:EUR&apikey=151J1CO4W4YLKP87'
r = requests.get(url)
data = r.json()

print(data)
```

```python
import pandas as pd

df = pd.DataFrame(data)
df.to_csv('sentiment.csv', index=False)
```
